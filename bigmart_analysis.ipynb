#  To Handle the Missing and Catagorical values

```python
# To handle the Misssing and null values
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
for col in train.columns:
  if train[col].dtype=='object':
    train[col]=le.fit_transform(train[col])

```

Purpose: Encodes categorical columns in the DataFrame train as numeric values. How It Works:
Initialization: Creates a LabelEncoder object.
Loop: Iterates over each column in train.
Check: If the column's data type is object (categorical), it transforms the values to numeric labels using
LabelEncoder.

```python
#Subset the dataset to include only the relevent columns
subset=train[['Item_Weight','Item_Visibility','Item_MRP',
'Outlet_Establishment_Year','Item_Outlet_Sales']] # Removed 'Item_Fat_Content' column
# Calculate the correlation matrix
corr=subset.corr()
# Create a mask for the upper triangle of the heatmap
mask=np.triu(np.ones_like(corr,dtype=bool))
# Set up the figure and axis objects
fig,ax=plt.subplots(figsize=(10,8))
# Generate the heatmap with green and white colors
sns.heatmap(corr,mask=mask,annot=True,cmap=sns.diverging_palette(20,220,n=200),
center=0,square=True,linewidths=.5,cbar_kws={"shrink":.5})
# Set the title and labels for the plot
ax.set_title("Correlation Heatmap",fontsize=18)
ax.set_xlabel("Features",fontsize=14)
ax.set_ylabel("Features",fontsize=14)
# Show the plot
plt.show()
```
